1) BASE DE DADOS

Este conjunto de dados "Estatísticas de Jogadores de Futebol", fornece uma visão detalhada das características e métricas dos atletas no futebol. Ele inclui informações como nome, nacionalidade, posição nacional e no clube, número da camisa, clubes e contratos, além de várias estatísticas de desempenho, como classificação, altura, peso, habilidades técnicas (como controle de bola, dribles e passes), e atributos físicos (como velocidade, força e agilidade).
```python
df.isnull().sum()
```

3) TRATAMENTO DOS DADOS FALTANTES

* Para valores numéricos, substitua por média/mediana:




```python
df['Contract_Expiry'].fillna(df['Contract_Expiry'].mean(), inplace=True)
df['National_Kit'].fillna(df['National_Kit'].mean(), inplace=True)
df['Club_Kit'].fillna(df['Club_Kit'].mean(), inplace=True)
df['Contract_Expiry'].fillna(df['Contract_Expiry'].mean(), inplace=True)
```

* Para valores categóricos, substitua por um valor padrão:

```python
df['National_Position'].fillna('Sub', inplace=True)
df['Club_Position'].fillna('Sub', inplace=True)
```

6) CRIAÇÃO DE GRÁFICOS (VISUALIZAÇÃO DOS DADOS)

**Gráficos para variáveis numéricas:**

* Histograma


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

top_5_alturas = df['Nationality'].value_counts().nlargest(5).index

df_top_5 = df[df['Nationality'].isin(top_5_alturas)]

plt.hist(df_top_5['Nationality'], bins=5, color='blue', edgecolor='black')
plt.title('Distribuição dos 5 Paises Mais Frequentes entre os Jogadores')
plt.xlabel('País')
plt.ylabel('Frequência')
plt.show()
```

* Boxplot

```python
top_5_alturas = df['Height'].value_counts().nlargest(5).index
df_top_5 = df[df['Height'].isin(top_5_alturas)]

sns.boxplot(x=df_top_5['Height'])
plt.title('Boxplot das 5 Alturas Mais Frequentes dos Jogadores')
plt.show()
```

**Gráficos para variáveis categóricas:**

* Gráfico de Barras


```python
df['Preffered_Foot'].value_counts().plot(kind='bar', color='green')
plt.title('Contagem de Jogadores por Preferência de Pé')
plt.xlabel('Lado')
plt.ylabel('Frequência')
plt.show()
```

* Gráfico de Pizza

```python
top_10_posicoes = df['Club_Position'].value_counts().nlargest(10)

top_10_posicoes.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporção de Jogadores por Posição (Top 10)')
plt.ylabel('')
plt.show()
```

**Gráficos relacionais:**

* Gráfico de Dispersão:


```python
top_10_clubes = df['Club'].value_counts().nlargest(10).index
df_top_10_clubes = df[df['Club'].isin(top_10_clubes)]


sns.scatterplot(x='Club', y='Rating', data=df_top_10_clubes)
plt.title('Avaliação do Jogador por Clube')
plt.xticks(rotation=90)
plt.show()
```
